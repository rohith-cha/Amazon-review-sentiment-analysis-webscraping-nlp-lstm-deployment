```python
import csv 
import pandas as pd
from bs4 import BeautifulSoup 
```

```python
#starting the webdriver
from selenium import webdriver
driver = webdriver.Chrome()
```

```python
url = 'https://www.amazon.in/New-Apple-iPhone-11-128GB/product-reviews/B08L8CXFZ1/ref=cm_cr_arp_d_viewopt_rvwer?ie=UTF8&reviewerType=avp_only_reviews&pageNumber=1'
reviewlist = []
```

```python
def get_soup(url):
    driver.get(url)    
    soup = BeautifulSoup(driver.page_source,'html.parser')
    return soup
```

```python
def get_reviews(soup):
    reviews = soup.find_all('div',{'data-hook':'review'})
    try:
        for item in reviews:
            review = {
            'product' : soup.title.text.replace('Amazon.in:Customer reviews:','').strip(),    
            'title' : item.find('a',{'data-hook': 'review-title'}).text.strip(),
            'rating' : item.find('i',{'data-hook':'review-star-rating'}).text.replace(' out of 5 stars','').strip(),
            'review_body' : item.find('span',{'data-hook': 'review-body'}).text.strip()    
            }
            reviewlist.append(review)
    except:
        pass
```

```python
get_soup(url)
```

```python
soup = get_soup('https://www.amazon.in/New-Apple-iPhone-11-128GB/product-reviews/B08L8CXFZ1/ref=cm_cr_arp_d_viewopt_rvwer?ie=UTF8&reviewerType=avp_only_reviews&pageNumber=1')
```

```python
get_reviews(soup)
```

```python
for x in range(1,501):
    soup = get_soup(f'https://www.amazon.in/New-Apple-iPhone-11-128GB/product-reviews/B08L8CXFZ1/ref=cm_cr_arp_d_viewopt_rvwer?ie=UTF8&reviewerType=avp_only_reviews&pageNumber={x}')
    print(f'Getting page: {x}')
    get_reviews(soup)
    print(len(reviewlist))
    if not soup.find('li',{'class': 'a-disabled a-last'}):
        pass
    else:
        break
```
